```
#read data into dataframe
df = pd.read_csv('/content/Speed_Dating_Data.csv', error_bad_lines=False, header='infer')
```

```
df.head()
```

```
#get the lowest and highest age of participants
print(df['age'].min())
print(df['age'].max())
```

```
#create new column with age difference
df['age_diff'] = abs(df['age']-df['age_o'])
df['age_diff'].head()
```

```
#create df with only matched couples
matched_df = df[df['match']==1]
```

```
matched_df.shape
```

```
#create df with only unmatched couples
unmatched_df = df[df['match']==0]
```

```
unmatched_df.shape
```

```
matched_df.describe()
```

```
unmatched_df.describe()
```

```
#columns to omit for exploratory visualization
omit_cols = ['iid', 'id', 'gender', 'idg', 'wave', 'round', 'position', 'positin1', 'order', 'partner', 'pid', 'match']
```

```
#create dataframes for matches by gender
matched_male = matched_df[matched_df['gender']==1]
matched_female = matched_df[matched_df['gender']==0]
```

```
#create data frames for unmatched by gender
unmatched_male = unmatched_df[unmatched_df['gender']==1]
unmatched_female = unmatched_df[unmatched_df['gender']==0]
```

```
#create the y values of matches for each gender
y_matched_male = [matched_male[i].mean() for i in matched_male.columns if matched_male[i].dtype!='object' and i not in omit_cols]
y_matched_female = [matched_female[i].mean() for i in matched_female.columns if matched_female[i].dtype!='object' and i not in omit_cols]
```

```
#create figure for plot
fig1 = go.Figure()

#create the plot for matched males
fig1.add_trace(go.Scatter(
    x=matched_male.columns,
    y=y_matched_male,
    name='Matched Male Interests',
    mode='markers',
    marker=dict(color='rgb(50, 50, 200)', size=10),
))

#create the plot for matched females
fig1.add_trace(go.Scatter(
    x=matched_female.columns,
    y=y_matched_female,
    name='Matched Female Interests',
    mode='markers',
    marker=dict(color='rgb(230, 50, 50)', size=10),
    marker_color='rgb(230, 50, 50)'))

#set any final attributes of the figure and display it
fig1.update_layout(plot_bgcolor='rgb(240,240,240)',
                  title='Popularity of Topic vs Matched Genders',
                  xaxis_title='Topic/Activity',
                  yaxis_title='Score')

fig1.show()
```

```
#write html file for importing into iframe on blog
io.write_html(fig1, 'topic_vs_matched_plot.html')
```

```
#create data for each gender who is not matched
y_unmatched_male = [unmatched_male[i].mean() for i in unmatched_male.columns if unmatched_male[i].dtype!='object' and i not in omit_cols]
y_unmatched_female = [unmatched_female[i].mean() for i in unmatched_female.columns if unmatched_female[i].dtype!='object' and i not in omit_cols]
```

```
#create the figure
fig2 = go.Figure()

#add data for unmatched males
fig2.add_trace(go.Scatter(
    x=unmatched_male.columns,
    y=y_unmatched_male,
    name='Matched Male Interests',
    mode='markers',
    marker=dict(color='rgb(50, 50, 200)', size=10),
))

#add data for unmatched females
fig2.add_trace(go.Scatter(
    x=unmatched_female.columns,
    y=y_unmatched_female,
    name='Matched Female Interests',
    mode='markers',
    marker=dict(color='rgb(230, 50, 50)', size=10),
    marker_color='rgb(230, 50, 50)'))

#perform final formatting on figure and display
fig2.update_layout(plot_bgcolor='rgb(240,240,240)',
                  title='Popularity of Topic vs Unmatched Genders',
                  xaxis_title='Topic/Activity',
                  yaxis_title='Score')
fig2.show()
```

```
#write html file for importing into iframe on blog
io.write_html(fig2, 'topic_vs_unmatched_plot.html')
```

```
#create dataframe organizing data by wave
waves_df = df[['wave', 'match', 'age_diff']]
```

```
#create list of waves for easy iteration
wave_list = sorted(waves_df['wave'].value_counts().index)
wave_list
```

```
#crosstab to better correlate age differences for each wave
age_diff_by_wave = pd.crosstab(index=df['age_diff'], columns=waves_df['wave'])
age_diff_by_wave
```

```
#number of matches for each wave
match_by_wave = pd.crosstab(index=waves_df['match'], columns=waves_df['wave'])
match_by_wave
```

```
#create figure
fig3 = go.Figure()

#add line data for mean age differences per wave
fig3.add_trace(go.Scatter(
    x=age_diff_by_wave.columns,
    y=[age_diff_by_wave[i].mean() for i in age_diff_by_wave.columns],
    name='Mean Age Difference Per Wave',
    mode='lines',
    marker=dict(color='rgb(245, 163, 0)', size=10),
))

#add data for average matches per wave
fig3.add_trace(go.Scatter(
    x=match_by_wave.columns,
    y=[(match_by_wave[i][1]) for i in match_by_wave.columns],
    name='Number of Matches Per Wave',
    mode='lines',
    marker=dict(color='rgb(50, 50, 200)', size=10),
))
#finish the figure and render the plot
fig3.update_layout(plot_bgcolor='rgb(247,247,247)',
                  title='Age Difference/Match Correlation by Wave',
                  xaxis_title='Wave',
                  yaxis_title='Mean Age Difference')
fig3.show()
```

```
#write html file for importing into iframe on blog
io.write_html(fig3, 'age_diff_vs_match_plot.html')
```

```
#dataframe for whether male or female is older
older_df = df[['wave', 'gender', 'age', 'age_o']]
older_df.head()
```

```
older_df.describe()
```

```
#create column with age difference relative to gender
#if negative the other gender is older
older_df['older'] = older_df['age']-older_df['age_o']
older_df.head()
```

```
#mean age difference
abs(older_df['older']).mean()
```

```
#separate by gender
older_female_df = older_df[older_df['gender'] == 0]
older_male_df = older_df[older_df['gender'] == 1]
```

```
older_female_df.head()
```

```
older_male_df.head()
```

```
#discern which person was older in each observation
temp_filter = older_female_df[older_female_df['older'] > 0]
```

```
temp_filter.head()
```

```
#crosstab of each age and total age difference for each wave in females
older_female_xtab = pd.crosstab(index=temp_filter['age'], columns=temp_filter['wave'])
older_female_xtab
```

```
#filter for males
temp_filter = older_male_df[older_male_df['older'] > 0]
```

```
temp_filter.head()
```

```
#crosstab of each age and total age difference for each wave in males
older_male_xtab = pd.crosstab(index=temp_filter['age'], columns=temp_filter['wave'])
older_male_xtab
```

```
#create the figure
fig4 = go.Figure()

#data for showing the mean age difference from the female perspective
fig4.add_trace(go.Scatter(
    x = older_female_xtab.columns,
    y = [older_female_xtab[i].mean() for i in older_female_xtab.columns],
    name = 'Female vs Male',
    mode = 'lines',
    marker = dict(color='rgb(240, 100, 105)')
))

#data for showing the mean age difference from the male perspective
fig4.add_trace(go.Scatter(
    x = older_male_xtab.columns,
    y = [older_male_xtab[i].mean() for i in older_male_xtab.columns],
    name = 'Male vs Female',
    mode = 'lines',
    marker = dict(color='rgb(105, 100, 230)')
))

#set figure attributes and render
fig4.update_layout(plot_bgcolor='rgb(247,247,247)',
                  title='Mean Age Difference by Gender',
                  xaxis_title='Wave',
                  yaxis_title='Mean Age Difference')

fig4.show()
```

```
#write html file for importing into iframe on blog
io.write_html(fig4, 'mean_age_by_gender_plot.html')
```
